```python
import pygsti
import pickle
```

```python
#Construct a target gateset: Identity, X(pi/2), Y(pi/2)
gs_target = pygsti.construction.build_gateset([2],[('Q0',)], ['Gi','Gx','Gy'], 
                             [ "I(Q0)","X(pi/2,Q0)", "Y(pi/2,Q0)"],
                             prepLabels=["rho0"], prepExpressions=["0"],
                             effectLabels=["E0"], effectExpressions=["1"], 
                             spamdefs={'plus': ('rho0','E0'), 'minus': ('rho0','remainder') } )

gs_target.set_all_parameterizations('TP')
```

```python
#Create gate string lists
fiducials = pygsti.construction.gatestring_list([ (), ('Gx',), ('Gy',), ('Gx','Gx'), ('Gx','Gx','Gx'), ('Gy','Gy','Gy') ]) # fiducials for 1Q MUB
fiducials_old = pygsti.construction.gatestring_list([ ('Gi',), ('Gx',), ('Gy',), ('Gx','Gx'), ('Gx','Gx','Gx'), ('Gy','Gy','Gy') ]) # Gi instead of {}

germs_all = pygsti.construction.gatestring_list( [
         ('Gx',), ('Gy',), ('Gi',), ('Gx', 'Gy'),
         ('Gx', 'Gy', 'Gi'), ('Gx', 'Gi', 'Gy'), ('Gx', 'Gi', 'Gi'), ('Gy', 'Gi', 'Gi'),
         ('Gx', 'Gx', 'Gi', 'Gy'), ('Gx', 'Gy', 'Gy', 'Gi'),
         ('Gx', 'Gx', 'Gy', 'Gx', 'Gy', 'Gy')] )

germs_wOmission = pygsti.construction.gatestring_list( [
                   ('Gx',), ('Gy',), ('Gi',),('Gx', 'Gy'),
                   ('Gx', 'Gy', 'Gi'), ('Gx', 'Gi', 'Gy'), ('Gx', 'Gi', 'Gi'),('Gy', 'Gi', 'Gi'),
                   ('Gx', 'Gx', 'Gi', 'Gy'),
                   ('Gx', 'Gx', 'Gy', 'Gx', 'Gy', 'Gy')] ) #Apparently ('Gx', 'Gy', 'Gy', 'Gi') is ommitted.

germs_onlyOneGi = pygsti.construction.gatestring_list( [
                    ('Gx',), ('Gy',), ('Gi',), ('Gx', 'Gy'),
                    ('Gx', 'Gx', 'Gy', 'Gx', 'Gy', 'Gy')] )

germs_noGi = pygsti.construction.gatestring_list( [('Gx',), ('Gy',), ('Gx', 'Gy'), ('Gx', 'Gx', 'Gy', 'Gx', 'Gy', 'Gy')] )

germs_old = pygsti.construction.gatestring_list( [('Gx',), ('Gy',), ('Gi',), ('Gx', 'Gy'), ('Gy','Gx')] )

maxLengths_8192 = [0,1,2,4,8,16,32,64,128,256,512,1024,2048,4096,8192]
maxLengths_4096 = [0,1,2,4,8,16,32,64,128,256,512,1024,2048,4096]
maxLengths_2048 = [0,1,2,4,8,16,32,64,128,256,512,1024,2048]
maxLengths_512 = [0,1,2,4,8,16,32,64,128,256,512]
maxLengths = [0,1,2,4,8,16,32,64,128,256,512]
maxLengths_old = [0,1,2,4,8,16,32,64,128,256]
```

```python
#Analysis of 04/17/2014 data.
result = pygsti.do_long_sequence_gst('../ExperimentalData/2014_04_17-compensatedTraining.txt',
                                     gs_target,fiducials_old,fiducials_old,germs_old,maxLengths_old,
                                     advancedOptions={'truncScheme':'length as exponent'})

pickle.dump(result, open("../ComputedObjects/2014_04_17-compensatedTraining.pkl", "w"))
```

```python
#Analysis of 12/02/2014
result = pygsti.do_long_sequence_gst('../ExperimentalData/2014_12_02-GSTCompensatedDrift_randomized_condensed.txt', 
                                     gs_target, fiducials, fiducials, germs_wOmission, maxLengths_512,
                                     advancedOptions={'truncScheme':'truncated germ powers'})

pickle.dump(result, open("../ComputedObjects/2014_12_02-GSTCompensatedDrift_randomized_condensed.pkl", "w"))
```

```python
#Analysis of 02/09/15
result = pygsti.do_long_sequence_gst('../ExperimentalData/2015_02_09-GSTCompensatedDrift_condensed.txt',
                                    gs_target, fiducials, fiducials, germs_all, maxLengths_512)

pickle.dump(result, open("../ComputedObjects/2015_02_09-GSTCompensatedDrift_condensed.pkl", "w"))
```
